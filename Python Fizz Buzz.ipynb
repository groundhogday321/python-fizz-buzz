# PYTHON FIZZ BUZZ

* loop through 100 numbers
* if divisible by 5 and 3 = Fizz Buzz
* if divisible by 3 = Fizz
* if divisible by 5 = Buzz
* else = num

# For Loop

```python
for i in range(1, 101):
    if i % 5 == 0 and i % 3 == 0:
        print(i, 'Fizz Buzz')
    elif i % 3 == 0:
        print(i, 'Fizz')
    elif i % 5 == 0:
        print(i, 'Buzz')
    else:
        print(i)
```

# For Loop Reverse

```python
for i in range(100, 0, -1):
    if i % 5 == 0 and i % 3 == 0:
        print(i, 'Fizz Buzz')
    elif i % 3 == 0:
        print(i, 'Fizz')
    elif i % 5 == 0:
        print(i, 'Buzz')
    else:
        print(i)
```

# While Loop

```python
i = 0
while i < 100:
    i += 1
    if i % 5 == 0 and i % 3 == 0:
        print(i, 'Fizz Buzz')
    elif i % 3 == 0:
        print(i, 'Fizz')
    elif i % 5 == 0:
        print(i, 'Buzz')
    else:
        print(i)
```

# While Loop Reverse

```python
i = 101
while i > 1:
    i -= 1
    if i % 5 == 0 and i % 3 == 0:
        print(i, 'Fizz Buzz')
    elif i % 3 == 0:
        print(i, 'Fizz')
    elif i % 5 == 0:
        print(i, 'Buzz')
    else:
        print(i)
```
